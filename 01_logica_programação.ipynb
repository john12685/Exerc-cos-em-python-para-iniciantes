## Tradução
####  print =  Impriir , input = Entrada , float = flutuante para números decimal, int = inteiro
####  if = se ,  elif = senão se ,  else = senão , and = E
 

```python
# O float() garante que o python aceite qualquer tipo de número 
num1 = float(input(("Digite o primeiro número: ")))
num2 = float(input("Digite o segundo número:" ))
num3 = float(input('Digite seu terceiro numero: '))

if(num1 >= num2) and (num1 >= num3):
    maior = num1
elif (num2 >= num1) and (num2 >= num3):
    maior = num2
else:
    maior = num3
print(f"O maior número é {maior}") 
```
